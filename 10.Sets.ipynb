# Set (Множества)

Множество в языке Питон — это структура данных, эквивалентная множествам в математике. Множество может состоять из различных элементов, порядок элементов в множестве неопределен. В множество можно добавлять и удалять элементы, можно перебирать элементы множества, можно выполнять операции над множествами (объединение, пересечение, разность). Можно проверять принадлежность элемента множеству.

В отличие от списка, где элементы хранятся в виде последовательного списка, в множествах порядок хранения элементов неопределен. Это позволяет выполнять операции типа “проверить принадлежность элемента множеству” быстрее, чем просто перебирая все элементы множества.

Элементами множества может быть любой неизменяемый тип данных: числа, строки, кортежи. Изменяемые типы данных не могут быть элементами множества, в частности, нельзя сделать элементом множества список (но можно сделать кортеж) или другое множество. Требование неизменяемости элементов множества накладывается особенностями представления множества в памяти компьютера.

# Задание множеств

Множество задается перечислением всех его элементов в фигурных скобках. Исключением явлеется пустое множество, которое можно создать при помощи функции set(). Если функции set передать в качестве параметра список, строку или кортеж, то она вернёт множество, составленное из элементов списка, строки, кортежа. Например:

```python
A = {1, 2, 3}
S = set('qwerty')
print(A)
print(S)
```

Каждый элемент может входить в множество только один раз, порядок задания элементов неважен. Например, программа ниже выведет True, так как A и B — равные множества.

```python
A = {1, 2, 3}
B = {3, 2, 3, 1}
print(A == B)
```

## Работа с элементами множеств


Узнать число элементов в множестве можно при помощи функции len.

Перебрать все элементы множества (в неопределенном порядке!) можно при помощи цикла for:

```python
primes = {2, 3, 5, 7, 11,2, 11}
for num in primes:
    print(num)
```

## Объединение множеств

В Python существует несколько способов соединения двух или более наборов.

Вы можете использовать union()метод, который возвращает новый набор, содержащий все элементы из обоих наборов, или update()метод, который вставляет все элементы из одного набора в другой:

```python
#Метод union()возвращает новый набор со всеми элементами из обоих наборов:
set1 = {"a", "b" , "c",1, 2}
set2 = {1, 2, 3}
setn = {7,9}

set3 = set2.union(set1,setn)
print(set3)
```

```python
#Метод update()вставляет элементы из set2 в set1:

set1 = {"a", "b" , "c"}
set2 = {1, 2, 3}

set1.update(set2)
print(set1)
```

## Удаление элементов из множества

Для удаления элемента x из множества есть два метода: discard и remove. Их поведение различается только в случае, когда удаляемый элемент отсутствует в множестве. В этом случае метод discard не делает ничего, а метод remove генерирует исключение KeyError.

Наконец, метод pop удаляет из множества один случайный элемент и возвращает его значение. Если же множество пусто, то генерируется исключение KeyError.

```python
fruits = {"apple", "banana", "cherry"}

fruits.remove("banana") #Попробуйте удалить элемент, которого нет в множестве

print(fruits)
```

```python
fruits = {"apple", "banana", "cherry"}

fruits.discard("a") #Попробуйте удалить элемент, которого нет в множестве

print(fruits)
```

```python
#Метод pop()удаляет случайный элемент из набора и возвращает в место вызова удаленный элемент
fruits = {"apple", "banana", "cherry"}
x = fruits.pop()
print(x)
```

### 2. Напишите программу, которая определяет количество различных символов, встречающихся в символьной строке

```python
def count_unique_characters(input_string):
    unique_characters = set(input_string)
    
    return len(unique_characters)

if __name__ == "__main__":
    user_input = input("Введите строку: ")
    unique_count = count_unique_characters(user_input)
    print(f"Количество различных символов: {unique_count}")
```

### 3. Напишите программу, которая находит все различные цифры в символьной строке

```python
def find_unique_digits(input_string):

    unique_digits = set()

    for char in input_string:
        if char.isdigit():
            unique_digits.add(char)

    return sorted(unique_digits)

if __name__ == "__main__":
    my_str = "Купили 6 яблок, 7 съели"
    unique_digits = find_unique_digits(my_str)
    
    print(f"Уникальные цифры в строке: {', '.join(unique_digits)}")
```

### 4. Напишите скрипт, который выводит набор всех символов, встречающиеся в двух заданных строках.

```python
def find_common_characters(str1, str2):
    set1 = set(str1)
    set2 = set(str2)
    
    common_characters = set1.intersection(set2)
    
    return sorted(common_characters)

if __name__ == "__main__":
    string1 = input("Введите первую строку: ")
    string2 = input("Введите вторую строку: ")
    
    common_chars = find_common_characters(string1, string2)
    
    print(f"Общие символы в обеих строках: {', '.join(common_chars)}")
```

### 5. Даны два списка чисел. Найдите все числа, которые входят как в первый, так и во второй список и выведите их в порядке возрастания

```python
def find_common_numbers(list1, list2):
    set1 = set(list1)
    set2 = set(list2)

    common_numbers = set1.intersection(set2)

    return sorted(common_numbers)

if __name__ == "__main__":

    list1 = [1, 2, 3, 4, 5, 6]
    list2 = [4, 5, 6, 7, 8, 9]
    
    common_nums = find_common_numbers(list1, list2)
    
    print(f"Общие числа в обоих списках: {common_nums}")
```

### 6.
Во входной строке записана последовательность чисел через пробел. Для каждого числа выведите слово YES (в отдельной строке), если это число ранее встречалось в последовательности или NO, если не встречалось.

```python
def check_numbers(sequence):
    seen_numbers = set()

    for number in sequence:
        if number in seen_numbers:
            print("YES")
        else:
            print("NO")
            seen_numbers.add(number)

if __name__ == "__main__":
    input_sequence = input("Введите последовательность чисел через пробел: ")
    
    numbers = input_sequence.split()
    
    check_numbers(numbers)
```

### 7.
Дан текст: в первой строке записано число строк, далее идут сами строки. Определите, сколько различных слов содержится в этом тексте.

```python
import string

def count_unique_words(text):
    text = text.translate(str.maketrans("", "", string.punctuation)).lower()
    
    words = text.split()
    
    unique_words = set(words)
    
    return len(unique_words)

if __name__ == "__main__":
    input_text = """4
She sells sea shells on the sea shore 
The shells that she sells are sea shells 
I'm sure So if she sells sea shells on the sea shore 
I'm sure that the shells are sea shore shells"""
    
    lines = input_text.strip().split('\n')[1:]
    

    full_text = ' '.join(lines)
    
    unique_word_count = count_unique_words(full_text)
    
    print(f"Количество различных слов в тексте: {unique_word_count}")
```
